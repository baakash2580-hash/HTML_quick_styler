3. Working of the Project:
In this project, an HTML code snippet is provided as input. The Python program uses the BeautifulSoup library to parse the HTML content using the HTML parser. After parsing, the HTML structure can be printed or displayed in a well-formatted way. The project also demonstrates how HTML content can be displayed in Google Colab using Python.

```python
from IPython.display import display, HTML

html_code = """
<html>
<head>
<title>Sample Page</title>
</head>

<body>

<h1>Hello World</h1>
<p>This is a simple paragraph</p>
<button>Click Me</button>

</body>
</html>
"""

display(HTML(html_code))
```

HTML Parser Create

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(html_code, "html.parser")

print(soup.prettify())
```

```python
# Style for heading
for h1 in soup.find_all("h1"):
    h1["style"] = "color:blue; text-align:center;"

# Style for paragraph
for p in soup.find_all("p"):
    p["style"] = "font-size:18px; color:green;"

# Style for button
for button in soup.find_all("button"):
    button["style"] = "background-color:red; color:white; padding:10px;"
```

```python
from IPython.display import display, HTML

display(HTML(str(soup)))
```
